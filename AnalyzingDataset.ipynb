The dataset might contain infinity values which should be substituted as they cannot be used effectively in calculations. For this purpose all columns containing infinity values are identified.

```python
df.columns = df.columns.str.lower().str.replace(' ', '_').str.replace('/','_')
df.columns
```

```python
df.replace('Infinity', 'inf')
```

```python
try:
    df['dst_port'] = df['dst_port'].astype(int)
except ValueError as e:
    invalid_literal = str(e).split(" ")[-1][:-1]
    print(f"Invalid literal: {invalid_literal}")
```

```python
df = df[pd.to_numeric(df['dst_port'], errors='coerce').notna()]
```

```python
df=df.astype(types)
```

```python
df.info(verbose=True, null_counts=True)
```

```python
inf_columns = [c for c in df.columns if df[df[c] == np.inf][c].count() > 0]

print(f'Columns containing infinity values {inf_columns}')
```

Hence, the two columns containing inf values are flow_byts_s and flow_pkts_s.
The inf values are replaced with nan values, subsequently imputing all nan values of the 2 columns with their mean.

```python
for col in inf_columns:
    df[col].replace([np.inf, -np.inf], np.nan, inplace=True)
    mean_value = df[col].mean()
    df[col].fillna(mean_value, inplace=True)
```

**Label Creation**

To analyze the dataset in terms of binary classification (benign/attack) and multi-classification (benign/attack-type) addtional label columns are created:

label_is_attack specifies if a network flow represents a benign or an malicious flow.

label_is_attack_[attack_type] specifies if a network flow represents a certain type of attack.



```python
df['label'] = df.label.astype('category')
df['label_code'] = df['label'].cat.codes
df['label_is_attack'] = df.label.apply(lambda x: 0 if x == 'Benign' else 1)
```

First, converted Label from object data type to category data type. Label column specifies different attack classes and using that we form label_code which takes integers for specifying different classses of attack.
The cat.codes method of the pandas.Categorical object is used to get the integer codes (index) of the categories in a categorical variable. It converts the categorical variable into numerical representation so that it can be used as a feature in machine learning models.

```python
attack_types = [a for a in df.label.value_counts().index.tolist() if a != 'Benign']

for a in attack_types:
    l = 'label_is_attack_' + a.replace('-', ' ').replace(' ', '_').lower()
    df[l] = df.label.apply(lambda x: 1 if x == a else 0)
```
